```python
#Import libraries for Data Analysis EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data
df = pd.read_csv('anime_data_cleaned.csv')
```

```python
#Rank the top 100 anime by ranking, popularity above 5, genre action and type TV anime 
df_top100 = df[(df['normalized_rating'] <= 100) & (df['popularity'] >= 5) & (df['type_tv'] == 'Anime')]
```

```python
df_top100 = df_top100.sort_values(by=['normalized_rating'], ascending=False)
df_top100.head(100)
```
